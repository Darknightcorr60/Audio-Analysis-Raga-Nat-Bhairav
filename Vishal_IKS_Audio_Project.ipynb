```python
from pydub import AudioSegment

# Yahan apni file ka naam check kar lena jo tumne upload ki hai
audio = AudioSegment.from_file("NAt bhairav sample.m4a", format="m4a")
audio.export("NAt bhairav sample.wav", format="wav")

print("Conversion Done! Ab hum 'NAt bhairav sample.wav' use kar sakte hain.")
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# 1. Ab hum nayi bani hui .wav file ko load karenge
file_path = "NAt bhairav sample.wav"
y, sr = librosa.load(file_path)

# 2. Graph ki size set karein
plt.figure(figsize=(15, 5))

# 3. Waveform banayein
librosa.display.waveshow(y, sr=sr, color='blue')

# 4. Graph par titles lagayein (Interview mein impression ke liye)
plt.title('Waveform: Raag Nat Bhairav Sample', fontsize=15)
plt.xlabel('Time (Seconds)')
plt.ylabel('Amplitude')

# 5. Graph dikhayein
plt.show()
```

```python
import numpy as np

# 1. Short-time Fourier Transform (STFT) - Ye awaaz ko frequencies mein todta hai
D = librosa.stft(y)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

# 2. Plotting ka setup
plt.figure(figsize=(15, 6))

# 3. Spectrogram dikhana
img = librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='log', cmap='magma')

# 4. Colorbar lagana (Taaki pata chale kahan sound intense hai)
plt.colorbar(img, format="%+2.0f dB")

plt.title('Spectrogram: Raag Nat Bhairav (Sound DNA)', fontsize=15)
plt.show()
```
